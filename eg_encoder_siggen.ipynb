```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def rectangular_wave(t, height=1.0, width=1.0, duty=0.5, phase=0.0):
    mod = (t + phase) % width
    return np.where(mod < duty * width, height, 0.0)
```

```python
t = np.linspace(0, 0.2, 1000000)
dA = rectangular_wave(t, height=5, width=0.002, duty=0.50, phase=0.0)
dB = rectangular_wave(t, height=5, width=0.002, duty=0.50, phase=0.0005)

fig, ax = plt.subplots(figsize=(10, 4), dpi=120)
ax.plot(t, dA, lw=1.0, color='red')
ax.plot(t, dB, lw=1.0, color='blue')
ax.set_xlim(0, 0.02)
ax.set_xlabel('Time [s]')
ax.set_ylabel('Amplitude')
```

```python
t = np.arange(0, 60, 1/100_000)
t = t.astype(np.float32)
dA = rectangular_wave(t, height=5, width=0.002, duty=0.50, phase=0.0)
dA = dA.astype(np.float32)
dB = rectangular_wave(t, height=5, width=0.002, duty=0.50, phase=0.0005)
dB = dB.astype(np.float32)
```

```python
np.savetxt('data/mock_signal.csv',np.array([t,dA,dB]).T, delimiter=',')
```
